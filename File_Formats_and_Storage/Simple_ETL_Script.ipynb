```python
import pandas as pd


def extract_data(file_path):
    data_df = pd.read_csv(file_path)
    return data_df


def rename_columns(data_df):
    data_df = data_df.drop(columns=["ID_y"])

    data_df = data_df.rename(columns={
        "ID_x": "order_row_id",
        "OrderID": "order_id",
        "Client_ID": "client_id",
        "Product_ID": "product_id",
        "QTY": "quantity",
        "DataStamp": "order_timestamp",
        "First": "first_name",
        "Last": "last_name",
        "Name": "product_name",
        "Unit(g)": "unit_grams",
        "Price": "price",
        "District": "district"
    })

    return data_df


def remove_missing_values(data_df):
    data_df = data_df.dropna(
        subset=["order_id", "client_id", "product_id", "quantity", "price"]
    )
    return data_df


def set_data_types(data_df):
    data_df["order_row_id"] = data_df["order_row_id"].astype(int)
    data_df["order_id"] = data_df["order_id"].astype(int)
    data_df["client_id"] = data_df["client_id"].astype(int)
    data_df["product_id"] = data_df["product_id"].astype(int)

    data_df["quantity"] = data_df["quantity"].astype(int)
    data_df["price"] = data_df["price"].astype(float)

    data_df["order_timestamp"] = pd.to_datetime(data_df["order_timestamp"])

    return data_df


def clean_text_columns(data_df):
    data_df["first_name"] = data_df["first_name"].astype(str).str.strip().str.title()
    data_df["last_name"] = data_df["last_name"].astype(str).str.strip().str.title()
    data_df["district"] = data_df["district"].astype(str).str.strip().str.title()
    data_df["product_name"] = data_df["product_name"].astype(str).str.strip().str.lower()

    return data_df


def filter_invalid_records(data_df):
    data_df = data_df[data_df["quantity"] > 0]
    data_df = data_df[data_df["price"] > 0]

    data_df = data_df.drop_duplicates()

    return data_df


def create_business_metrics(data_df):
    data_df["revenue"] = data_df["quantity"] * data_df["price"]
    return data_df


def load_data(data_df, output_path):
    data_df.to_csv(output_path, index=False)


data_df = extract_data("/Users/fredikamionka/PycharmProjects/Projects_for_CV/Files_Project_for_CV/data.csv")
data_df = rename_columns(data_df)
data_df = remove_missing_values(data_df)
data_df = set_data_types(data_df)
data_df = clean_text_columns(data_df)
data_df = filter_invalid_records(data_df)
data_df = create_business_metrics(data_df)

load_data(data_df, "/Users/fredikamionka/PycharmProjects/Projects_for_CV/Files_Project_for_CV/clean_data.csv")
```
